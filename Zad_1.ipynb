```
import pandas as pd
import numpy as np
from sklearn import preprocessing

data = {
    'size' : ['L', 'L', 'M', 'XL', 'M', 'S', 'M', 'XXL', 'XL', 'L', 'M', 'XL', 'S', 'S', 'XL'],
    'color' : ['green', 'blue', 'red', 'blue', 'green', 'black', 'white', 'yellow', 'blue', 'black', 'green', 'blue', 'white', 'black', 'yellow'],
    'gender' : ['male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male'],
    'price' : [149.0, 80.0, 110.0, 121.0, 130.0, 74.0, 97.0, 114.0, 190.0, 230.0, 88.0, 114.0, 131.0, 165.0, 191.0],
    'weight' : [380, 520, 160, 520, 370, 266, 410, 420, 440, 610, 330, 250, 300, 310, 211],
    'bought': ['no', 'yes', 'no', 'yes', 'no', 'yes', 'no', 'yes', 'no', 'yes', 'no', 'yes', 'no', 'yes', 'no']
}

df_raw = pd.DataFrame(data=data)
df = df_raw.copy()

for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

for col2 in ['weight']:
  df[col2] = df[col2].astype('float')

df.describe(include=['category']).T

le = preprocessing.LabelEncoder()
df['bought'] = le.fit_transform(df['bought'])

pd.get_dummies(data=df, drop_first=True)

scaler = preprocessing.StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

df_raw
```
