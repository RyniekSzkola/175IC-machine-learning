```
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.datasets import make_blobs
```

```
data, _ = make_blobs(n_samples = 1000, centers = 4, random_state= 98, cluster_std = 1.4)
df = pd.DataFrame(data, columns = ['x1','x2'])
px.scatter(df,'x1','x2')
```

```
from sklearn.cluster import DBSCAN

cluster = DBSCAN(eps = 0.45, min_samples = 5)
cluster.fit(data)
df['cluster'] = cluster.labels_
px.scatter(df, 'x1', 'x2', 'cluster', template='plotly_dark', color_continuous_midpoint = 0)
```

```
from sklearn.cluster import KMeans

k_means_cluster = KMeans(n_clusters=4,random_state=98)
k_means_cluster.fit(df)
df['k_means'] = k_means_cluster.labels_
px.scatter(df, 'x1', 'x2', 'k_means', template='plotly_dark', color_continuous_midpoint = 0)
```

```
import pandas as pd

dbscan = df['cluster'].value_counts()[1:5].sort_index()
#Liczba punktw przypisanych do poszczeglnych klastrw przez algorytm DBSCAN
print(dbscan)
kmeans = df['k_means'].value_counts().sort_index()
#Liczba punktw dla algorytmu KMEANS
print(kmeans)
```
